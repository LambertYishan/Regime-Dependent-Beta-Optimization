```python
"""
EDA Visuals for Factor and Equity Data
======================================

This script performs exploratory data analysis (EDA) on two Excel workbooks
provided alongside this script:

* ``Factor Equities.xlsx`` – a panel of monthly closing prices for roughly
  232 different equities. Each row corresponds to a month-end date (in
  the ``Dates`` column) and each subsequent column holds the price for a
  particular equity.
* ``Master Excel Lambert.xlsx`` – a macro factors data set with monthly
  observations for various economic and market indicators. The sheet
  ``Sheet1`` contains columns such as ``Low High Yield OAS`` (credit
  spreads), ``NonFarm ManufacturingWeekly hours``, ``30Y T‑Yield``, etc.

The goal of this script is to surface insightful patterns by computing
returns, correlations and covariances, visualising distributions and
relationships, and inspecting how macro factors relate to an aggregated
equity return. Charts are saved as individual PNG files in the working
directory. To run this script you will need the pandas and matplotlib
packages available in your environment.

Usage::

    python eda_visuals.py

The script assumes that both Excel files reside in the same directory as
``eda_visuals.py``. If they are located elsewhere, update the paths
accordingly at the top of the script.
"""
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib
# Use a non-interactive backend to ensure plotting works in headless
# environments. Without this line matplotlib attempts to call an internal
# service exposed by Jupyter, which can cause connection errors when no
# notebook server is available. The 'Agg' backend writes plots directly
# to image files.
matplotlib.use("Agg")
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Configuration
# -----------------------------------------------------------------------------
# Define file names. Adjust these if your files reside elsewhere or are named
# differently.
FACTOR_FILE = "Factor Equities.xlsx"
MACRO_FILE = "Master Excel Lambert.xlsx"

# Output directory for plots. Defaults to the current directory. Create it
# if it doesn't exist.
OUTPUT_DIR = os.getcwd()
os.makedirs(OUTPUT_DIR, exist_ok=True)

```

```python

def load_factor_equities(path: str) -> pd.DataFrame:
    """Load the equities price data and set the index to datetime.

    Parameters
    ----------
    path : str
        Path to the Excel file containing the ``Equities`` sheet.

    Returns
    -------
    pd.DataFrame
        DataFrame indexed by month-end dates with price columns for each
        equity.
    """
    df = pd.read_excel(path, sheet_name="Equities", parse_dates=["Dates"])
    # Ensure the date index is at month end to aid in merging with macro data.
    df["Dates"] = df["Dates"].dt.to_period("M").dt.to_timestamp("M")
    return df.set_index("Dates")


def load_macro_factors(path: str) -> pd.DataFrame:
    """Load the macro factors and set the index to month end dates.

    Parameters
    ----------
    path : str
        Path to the Excel file containing the ``Sheet1`` sheet.

    Returns
    -------
    pd.DataFrame
        DataFrame indexed by month-end dates with various macro indicators.
    """
    df = pd.read_excel(path, sheet_name="Sheet1", parse_dates=["observation_date"])
    # Convert observation_date to period month and then to month end timestamp
    df["observation_date"] = df["observation_date"].dt.to_period("M").dt.to_timestamp("M")
    return df.set_index("observation_date")


def compute_equity_returns(price_df: pd.DataFrame) -> pd.DataFrame:
    """Compute simple percentage change for each equity column.

    The first row will be dropped because pct_change produces NaN for the
    initial observation.

    Parameters
    ----------
    price_df : pd.DataFrame
        A DataFrame indexed by dates with price columns.

    Returns
    -------
    pd.DataFrame
        DataFrame of returns on the same index. Columns correspond to
        equities. Values are decimal returns (e.g. 0.02 = 2% return).
    """
    return price_df.pct_change().dropna(how="all")


def plot_corr_heatmap(matrix: pd.DataFrame, title: str, filename: str, labels: list | None = None):
    """Plot a correlation or covariance heatmap using matplotlib.

    Parameters
    ----------
    matrix : pd.DataFrame
        Square correlation or covariance matrix.
    title : str
        Title for the plot.
    filename : str
        Filename (without directory) to save the plot as a PNG.
    labels : list, optional
        Label names for the axes. If None, DataFrame column names are used.
    """
    plt.figure(figsize=(10, 8))
    # Use imshow for the heatmap. The default colormap is used to avoid
    # specifying colours explicitly.
    plt.imshow(matrix, interpolation='nearest')
    plt.colorbar()
    if labels is None:
        labels = matrix.columns.tolist()
    plt.xticks(range(len(labels)), labels, rotation=90, fontsize=6)
    plt.yticks(range(len(labels)), labels, fontsize=6)
    plt.title(title)
    plt.tight_layout()
    out_path = os.path.join(OUTPUT_DIR, filename)
    plt.savefig(out_path, dpi=300)
    plt.close()
```

```python

def main():
    # -------------------------------------------------------------------------
    # Load data
    # -------------------------------------------------------------------------
    factor_path = os.path.join(OUTPUT_DIR, FACTOR_FILE)
    macro_path = os.path.join(OUTPUT_DIR, MACRO_FILE)
    if not os.path.exists(factor_path) or not os.path.exists(macro_path):
        raise FileNotFoundError(
            f"Could not locate '{FACTOR_FILE}' or '{MACRO_FILE}'. "
            f"Ensure they reside in {OUTPUT_DIR} or update the paths in the script."
        )

    # Read the data
    print("Loading equity prices...")
    price_df = load_factor_equities(factor_path)
    print(f"Loaded equity prices with shape {price_df.shape}")

    print("Loading macro factors...")
    macro_df = load_macro_factors(macro_path)
    print(f"Loaded macro data with shape {macro_df.shape}")

    # -------------------------------------------------------------------------
    # Compute and visualise basic statistics
    # -------------------------------------------------------------------------
    # Compute monthly returns for equities
    returns_df = compute_equity_returns(price_df)
    print(f"Computed returns with shape {returns_df.shape}")

    # Compute a simple equal-weighted average return across all equities to
    # approximate the overall market. This is used later to compare with macro
    # factors.
    market_return = returns_df.mean(axis=1).rename("Market_Return")

    # -------------------------------------------------------------------------
    # Distribution of returns
    # -------------------------------------------------------------------------
    print("Plotting distribution of monthly equity returns...")
    # Plot distribution of all monthly equity returns using matplotlib directly
    values = returns_df.values.flatten()
    values = values[~np.isnan(values)]
    plt.figure(figsize=(8, 6))
    plt.hist(values, bins=100)
    plt.title("Distribution of Monthly Equity Returns")
    plt.xlabel("Return")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "return_distribution.png"), dpi=300)
    plt.close()

    # -------------------------------------------------------------------------
    # Correlation matrix for a subset of equities
    # -------------------------------------------------------------------------
    # To keep the correlation heatmap readable, select the first 10 equities.
    # Feel free to adjust 'n_equities' to a larger number, but note that
    # readability suffers as the matrix grows.
    n_equities = 10
    sample_cols = returns_df.columns[:n_equities]
    print(f"Computing correlation matrix for {n_equities} equities...")
    sample_corr = returns_df[sample_cols].corr()
    plot_corr_heatmap(
        sample_corr,
        title="Correlation Matrix – Sample of Equity Returns",
        filename="equities_sample_corr_heatmap.png",
        labels=sample_cols.tolist(),
    )

    # -------------------------------------------------------------------------
    # Correlation and covariance matrices for macro factors
    # -------------------------------------------------------------------------
    # Exclude any non-numeric columns (the index is not included by default).
    macro_numeric = macro_df.select_dtypes(include=[np.number])
    print("Computing macro factor correlation matrix...")
    macro_corr = macro_numeric.corr()
    plot_corr_heatmap(
        macro_corr,
        title="Correlation Matrix – Macro Factors",
        filename="macro_corr_heatmap.png",
        labels=macro_corr.columns.tolist(),
    )

    print("Computing macro factor covariance matrix...")
    macro_cov = macro_numeric.cov()
    plot_corr_heatmap(
        macro_cov,
        title="Covariance Matrix – Macro Factors",
        filename="macro_cov_heatmap.png",
        labels=macro_cov.columns.tolist(),
    )

    # -------------------------------------------------------------------------
    # Relationship between macro factors and market return
    # -------------------------------------------------------------------------
    print("Computing correlation between macro factors and market return...")
    # Align macro data with the market return on month end. Because both data
    # sets are indexed by month end, joining on the index automatically aligns
    # the dates. Outer join is avoided to ensure correlations are calculated
    # only when both series have data.
    combined = macro_numeric.join(market_return, how="inner")
    # Drop rows where the market return or macro factors are NaN
    combined_clean = combined.dropna(how="any")
    # Compute correlations of each macro factor with the market return
    macro_market_corr = combined_clean.corr()["Market_Return"].drop("Market_Return")
    # Bar chart of macro correlations with market return
    # Plot bar chart of macro factor correlations with the market return using
    # matplotlib directly to avoid pandas' built-in plotting hooks.
    indices = np.arange(len(macro_market_corr))
    plt.figure(figsize=(10, 6))
    plt.bar(indices, macro_market_corr.values)
    plt.xticks(indices, macro_market_corr.index, rotation=90)
    plt.title("Correlation of Macro Factors with Average Equity Return")
    plt.ylabel("Correlation coefficient")
    plt.xlabel("Macro Factor")
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "macro_vs_market_correlation.png"), dpi=300)
    plt.close()

    # -------------------------------------------------------------------------
    # Scatter plots of selected macro factors vs market return
    # -------------------------------------------------------------------------
    # Some macro variables are likely to show relationships with returns. Here
    # we choose a handful of factors for inspection. Adjust this list as
    # desired to explore more combinations.
    selected_macros = [
        "Low High Yield OAS",
        "30Y T-Yield",
        "MSCI Low Size",
        "BCOM",
        "CBOE Total Put-Call Ratio",
    ]
    for col in selected_macros:
        if col not in combined_clean.columns:
            print(f"Warning: macro factor '{col}' not found in data; skipping plot.")
            continue
        plt.figure(figsize=(8, 6))
        plt.scatter(combined_clean[col], combined_clean["Market_Return"])
        plt.xlabel(col)
        plt.ylabel("Market Return")
        plt.title(f"{col} vs Market Return")
        plt.tight_layout()
        safe_name = col.replace(" ", "_").replace("/", "_")
        plt.savefig(
            os.path.join(OUTPUT_DIR, f"{safe_name}_vs_market_return_scatter.png"),
            dpi=300,
        )
        plt.close()

    # Inform the user of where charts have been saved
    print(f"EDA complete. Plots are saved in '{OUTPUT_DIR}'.")


if __name__ == "__main__":
    main()
```
